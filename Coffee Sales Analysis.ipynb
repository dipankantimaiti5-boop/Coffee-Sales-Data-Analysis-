```python


import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 


data=pd.read_csv('e:\Sree Sree Thakur   ( MORPHIS ) And Dipan Kanti Maiti ( NEO ) And Trinity ( IUSEE )\Data Analyst Intern, 2026 @ Unified Mentor\Project 1\index_csv.csv')
data.head() # Displaying the first few rows 

data.shape # Getting the dimension of the data frame 
data.index # Getting the row names of the data frame 

data.isnull().sum() # checking for null values / missing values 

sns.displot(data['card']) # Distribution plot to deal with null values of card column 

data[data.duplicated()] # Duplicates 
data.sort_values(by='coffee_name') # sorting by default asending order 

coffee=['Latte','Hot Chocolate','Americano','Americano with Milk','Cocoa']
Sales=[38.7,33.8,28.9,40,24]
plt.pie(Sales,labels=coffee,colors=['r','g','b','c','y'],autopct='%0.1f%%') #pie ploting method 
plt.show()

coffee_name=['lt','hc','ac','am','ca']
Sales=[38.7,33.8,28.9,40,24]
index=np.arange(len(coffee_name))
plt.bar(index,Sales,width=0.2,color='c',edgecolor='r',linewidth=2) #bar ploting method
plt.xticks(index+0.2,coffee_name)
plt.title("coffee wise sales")
plt.xlabel('coffee_name')
plt.ylabel('money')
plt.show()






 



```

## conclusion: We see that on Pie chart cocoa with 24 dollar sales very less but americano with milk with 40 dollar sales very high and latte with 38.7 dollar is close to americano with milk 

##conclusion: we saw that on bar plot am i.e. americano with milk coffee sales higher than other types of coffee and ca i.e. cocoa sales lower than other types of coffee , however latte coffee sales is close to americano milk coffee and hot chocolate and americano also sales close to each other.
